```python
Q16. Create a summary report that includes the total sales, average price, and total
quantity sold for each product class.
```

```python
Q17. Find the top 5 customers with the highest sales for each year
```

```python
data.groupby([data['Year'], 'Customer Name'])['Sales'].sum().groupby(level=0, group_keys=False).nlargest(5)
```

```python
Q18. Calculate the year-over-year growth in sales for each country.
```

```python
data.groupby(['Country', data.Year])['Sales'].sum().unstack().pct_change(axis=1)
```

```python
19. List the months with the lowest sales for each year
```

```python
data.groupby([data.Year, data.Month])['Sales'].sum().groupby(level=0, group_keys=False).nsmallest(1)
```
